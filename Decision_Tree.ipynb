1. What are some common hyperparameters of decision tree models, and how do they affect the model's performance?

Decision tree models have several hyperparameters that directly affect their complexity, accuracy, and ability to generalize. Here's a list of some common hyperparameters of decision tree models and their impact on model performance:
1. Max Depth (max_depth)
Description: The maximum depth of the tree (i.e., the longest path from the root to a leaf)
2. Minimum Samples Split (min_samples_split)
Description: The minimum number of samples required to split an internal node.
3. Minimum Samples Leaf (min_samples_leaf)
Description: The minimum number of samples required to be at a leaf node.


. What is the difference between the Label encoding and One-hot encoding?


Label Encoding and One-Hot Encoding are both techniques used to convert categorical variables into numerical values, but they do so in different ways. Here’s a breakdown of the key differences between the two:
Definition: Label encoding assigns a unique integer value to each category of a categorical variable
Definition: One-hot encoding creates a new binary column (or feature) for each possible category, where each column represents one category, and the value is 1 if the observation belongs to that category, and 0 if it does not.
